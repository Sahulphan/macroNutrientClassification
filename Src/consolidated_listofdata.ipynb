```python
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import numpy as np
import pandas as pd
from __future__ import division
from math import sqrt
import os
import fnmatch
import glob
import csv
import statistics
import calendar
from datetime import datetime
from sklearn import metrics
```

```python
mealid_list=[]
Featuremean_list = []
Featurestd_list = []
Featureauc_list = []
Featurepeak_list = []
maxindextime_list = []
bwtime_list = []
Featureskew_list = []
FeatureLastHHMean_list = []
fat_list = []
carb_list = []
protein_list = []
truncatedID_list = []
df=pd.DataFrame()
for csvfile in glob.glob(os.path.join('/home/sahulphaniraj/csce633-ml/MachineLearningProject/Dataset/PlosJournalData/Consolidated-features/', '*.csv')):
    mealid_withspace = csvfile[100:]
    mealid=''
    for i in range(len(mealid_withspace)):
        char = mealid_withspace[i]
        if(char.isspace() == False):
            mealid+=char
    mealid=mealid.replace('.csv','',1)
    mealid_list.append(mealid)
    #read featuresfrom file
    filename = csvfile
    csv_input = pd.read_csv(filename)
    Featuremean_list.append(csv_input['FeatureMean'][0])
    Featurestd_list.append(csv_input['FeatureStd'][0])
    Featureauc_list.append(csv_input['FeatureAUC'][0])
    Featurepeak_list.append(csv_input['FeaturePeak'][0])
    maxindextime_list.append(csv_input['MaxIndex'][0])
    bwtime_list.append(csv_input['BandwidthTime'][0])
    Featureskew_list.append(csv_input['FeatureSkew'][0])
    FeatureLastHHMean_list.append(csv_input['FeatureLastHHMean'][0])
    #ground truth values
    if (mealid.find("PB") != -1):
        fat_list.append(20)
        carb_list.append(51)
        protein_list.append(18)
    elif(mealid.find("Bar") != -1):
        fat_list.append(18)
        carb_list.append(48)
        protein_list.append(9)  
    elif(mealid.find("CF") != -1):
        fat_list.append(2.5)
        carb_list.append(54)
        protein_list.append(11) 
    #truncated ID list
    truncatedID_list.append(mealid[0:8])
df['ID'] = mealid_list  
df['TruncatedID'] = truncatedID_list
df['Mean'] = Featuremean_list
df['SD'] = Featurestd_list
df['AUC'] = Featureauc_list
df['Peak'] = Featurepeak_list
df['MaxIndex'] = maxindextime_list
df['Halfpk'] = bwtime_list
df['Skew'] = Featureskew_list
df['HHMean'] = FeatureLastHHMean_list
df['Fat'] = fat_list
df['Carb'] = carb_list
df['Protein'] = protein_list
df.to_csv('/home/sahulphaniraj/csce633-ml/MachineLearningProject/Dataset/consolidated_list.csv', index=False) #write to file
```
