```python
CgmDataSorted = CgmData.sort_values(['userID','Meal'],ascending=[True,True])
```

```python
CgmDataSorted = CgmDataSorted.reset_index(drop=True)
```

```python
CgmDataSorted['GlucoseValue']
```

```python
len(CgmData['userID'].unique())
```

```python
userlist = CgmDataSorted['userID'].unique()
mealist = ['PB','CF','Bar']
for user in range(len(userlist)):
    for meal in range(1):
        username = userlist[user]
        mealname = mealist[meal]
        temp=CgmDataSorted.loc[(CgmDataSorted['userID']==username) & (CgmDataSorted['Meal']==mealname)]
        filename = 'User'+user+'Meal'+meal
        temp.to_csv(filename,sep='\t')
        
        
```
